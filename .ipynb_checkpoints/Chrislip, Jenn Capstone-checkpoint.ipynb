```python
# find the percentage of employed Americans working from home or teleworking at least part of the time.

perc_of_WFH_by_industry
```

```python
# drop columns with NaN

perc_of_WFH_by_industry = perc_of_WFH_by_industry.dropna().sort_values('Industry').reset_index(drop=True)

perc_of_WFH_by_industry
```

```python
perc_of_WFH_by_industry.info()
```

```python
# Calculate the different between 2019 and 2021

perc_of_WFH_by_industry['2019_vs_2021 Change'] = perc_of_WFH_by_industry[2021] - perc_of_WFH_by_industry[2019]
```

```python
# Calculate the different between 2019 and 2021

perc_of_WFH_by_industry['2021_vs_2022 Change'] = perc_of_WFH_by_industry[2022] - perc_of_WFH_by_industry[2021]
```

```python
# Separate the new columns for graphing purposes & sort values in ascending order

perc_of_WFH_2019_2021 = perc_of_WFH_by_industry[['Industry', '2019_vs_2021 Change']].sort_values('2019_vs_2021 Change')
```

```python
# convert df to csv

perc_of_WFH_by_industry.to_csv('perc_of_WFH_by_industry.csv')
```

```python
# Create Horizontal Bar Graph

plt.barh(perc_of_WFH_2019_2021['Industry'],perc_of_WFH_2019_2021['2019_vs_2021 Change'])
plt.xlabel('Growth Percentage')
plt.title('Growth of Remote Workers by Industry from 2019 to 2021')
plt.show();
```

```python
# Separate the new columns for graphing purposes & sort values in ascending order

perc_of_WFH_2021_2022 = perc_of_WFH_by_industry[['Industry', '2021_vs_2022 Change']].sort_values('2021_vs_2022 Change')
```

```python
# Create Horizontal Bar Graph

plt.barh(perc_of_WFH_2021_2022['Industry'],perc_of_WFH_2021_2022['2021_vs_2022 Change'])
plt.xlabel('Decline Percentage')
plt.title('Decline of Remote Workers by Industry from 2021 to 2022')
plt.show();
```

```python
gdp = gdp_2019_2024.loc[gdp_2019_2024['Country Name'] == 'United States']

US_gdp = gdp[['Country Name','Indicator Name', '2019', '2020', '2021', '2022', '2023', '2024']]

# US_gdp

US_gdp
```

```python
US_gdp.info()
```

```python
# Show full numbers on GDP rather than scientific notation

US_gdp['2019'] = pd.to_numeric(US_gdp['2019'], downcast='integer', errors='coerce')
```

```python
US_gdp
```

```python
# Convert $$ amounts to short-hand in trillions for easier readability

cols = ['2019','2020','2021','2022','2023','2024']
df_trillions = US_gdp.copy()
df_trillions[cols] = US_gdp[cols] / 1_000_000_000_000

df_trillions
```

```python
# Limit decimal to 3 places

df_trillions[cols] = df_trillions[cols].applymap(lambda x: f"{x:.3f}")
```

```python
df_trillions
```

```python
# Convert GDP df to csv

df_trillions.to_csv('df_trillions.csv')
```

```python
# Pull Year out of date and put into its own column

monthly_pop_2010_2020['Year'] = monthly_pop_2010_2020['Year and Month'].dt.year

monthly_pop_2010_2020
```

```python
# Pull Year out of date and put into its own column

monthly_pop_2020_2025['Year'] = monthly_pop_2020_2025['Year and Month'].dt.year

monthly_pop_2020_2025
```

```python
# Group the population by year and sum together

resi_pop1 = monthly_pop_2020_2025.groupby('Year')['Resident Population Plus Armed Forces Overseas'].sum().to_frame()

resi_pop1
```

```python
# Group the population by year and sum together

resi_pop2 = monthly_pop_2010_2020.groupby('Year')['Resident Population Plus Armed Forces Overseas'].sum().to_frame()

resi_pop2
```

```python
# Merge the two Residential population df's together into one df

resi_pop = pd.concat([resi_pop1, resi_pop2]).sort_values(by='Year')

resi_pop
```

```python
resi_pop.to_csv('residential_and_armed_forces_population.csv')
```
